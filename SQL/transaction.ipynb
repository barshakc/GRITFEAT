```python
import pandas as pd 
import numpy as np 

df=pd.read_csv("C:\\Transactions.csv")
df.head()
```

```python
df.shape
```

```python
# Check for duplicates
duplicate_ids = df[df.duplicated(subset='transaction_id', keep=False)]['transaction_id'].unique()

print("Duplicate transaction IDs:")
print(duplicate_ids)

print(f"Total duplicate transaction IDs: {len(duplicate_ids)}")
```

```python
df_cleaned = df.drop_duplicates(subset='transaction_id', keep='first')
print(f"Rows after removing duplicates: {len(df_cleaned)}")
```

```python
df_cleaned.to_csv("transactions_cleaned.csv", index=False)
```

```python
# Check missing values count per column
missing_values = df_cleaned.isnull().sum()

print("Missing values per column:")
print(missing_values)
```

```python
df1= pd.read_csv("transactions_cleaned.csv")
```

```python
# Convert tran_date from DD-MM-YYYY to YYYY-MM-DD
df1['tran_date'] = pd.to_datetime(df1['tran_date'], dayfirst=True, errors='coerce')
print(df1['tran_date'].head())

# Convert back to string for MySQL import
df1['tran_date'] = df1['tran_date'].dt.strftime('%Y-%m-%d')
print(df1['tran_date'].head())

```

```python
missing_tran_date = df1['tran_date'].isnull().sum()
print(f"Number of missing or null values in 'tran_date': {missing_tran_date}")
```

```python
#Fill missing values with default date
df1['tran_date'].fillna('2000-01-01', inplace=True)
```

```python
missing_tran_date = df1['tran_date'].isnull().sum()
print(missing_tran_date)
```

```python
missing_values = df1.isnull().sum()
print("Missing values per column:")
print(missing_values)
```

```python
df1.to_csv("transaction.csv", index=False)
```
