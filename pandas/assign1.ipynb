DATA TRANSFORMATION

```python
# Use .apply() to categorize data
def categorize_gpa(gpa):
    if gpa >= 3.5:
        return 'Excellent'
    elif gpa >= 3.0:
        return 'Good'
    else:
        return 'Needs Improvement'

df['GPA_Category'] = df['GPA'].apply(categorize_gpa)
df.head()
```

```python
# 2. Use .map() to convert data
gender_map = {'Male': 'M', 'Female': 'F', 'Other': 'O'}
df['Gender_Code'] = df['gender'].map(gender_map)
```

```python
# Create a new column using vectorized operations
# let new column be total activity
df['Total_Activity'] = df['lms_logins_past_month'] + df['forum_participation_count']
```

```python
# Conditional Column  Two Values Based on Condition
df['Risk_Flag'] = df['risk_level'].apply(lambda x: 'Alert' if x == 'High' else 'Normal')
df.head()
```

Data Cleaning 

```python
# 1. Count missing values in each column
missing_counts = df.isnull().sum()
print(missing_counts)
```

Since we dont have any missing values, lets introduce some to explore techniques of filling missing values.

```python
df.loc[0, 'Grade Point Average'] = np.nan
df.loc[5, 'Major'] = np.nan
df.loc[10, 'attendance_rate'] = np.nan
print(df.isnull().sum())
```

```python
# Replace missing GPA with the mean
gpa_mean = df['Grade Point Average'].mean()
df['Grade Point Average'].fillna(gpa_mean, inplace=True)
```

```python
# Replace missing attendance_rate with the median
attendance_median = df['attendance_rate'].median()
df['attendance_rate'].fillna(attendance_median, inplace=True)
```

```python
#Replace missing major with mode
major_mode = df['Major'].mode()[0]
df['Major'].fillna(major_mode, inplace=True)
```

```python
print(df.isnull().sum())
df.head()
```

```python
#Drop duplicates 
df.drop_duplicates(inplace=True)
```

```python
# Clean 'Major' column by stripping whitespace and converting to title case
df['Major'] = df['Major'].str.strip().str.title()
```

Memory and Index Optimization

```python
# 1. Inspect memory usage
df.memory_usage(deep=True)
```

```python
# Initial memory usage
initial_memory = df.memory_usage(deep=True)
print(f"Initial Memory Usage:\n{initial_memory.sum() / 1024**2:.2f} MB")

# Set student_id as index and compare memory usage
df_indexed = df.copy().set_index('student_id')
indexed_memory = df_indexed.memory_usage(deep=True)
print(f"\nMemory Usage after setting index:\n{indexed_memory.sum() / 1024**2:.2f} MB")
```

```python
# Convert string/object columns to 'category' dtype
df_category = df.copy()
for col in ['gender', 'major', 'enrollment_status', 'risk_level']:
    df_category[col] = df_category[col].astype('category')
category_memory = df_category.memory_usage(deep=True)
print(f"\nMemory Usage after converting object columns to category:\n{category_memory.sum() / 1024**2:.2f} MB")
print(f"Memory saved: {(initial_memory.sum() - category_memory.sum()) / 1024**2:.2f} MB")
```

```python
# Convert float64 to float32
df_float32 = df_category.copy()
for col in df_float32.select_dtypes(include='float64').columns:
    df_float32[col] = df_float32[col].astype('float32')
float32_memory = df_float32.memory_usage(deep=True)
print(f"\nMemory Usage after converting float64 to float32:\n{float32_memory.sum() / 1024**2:.2f} MB")
print(f"Memory saved: {(category_memory.sum() - float32_memory.sum()) / 1024**2:.2f} MB")
```

Vectorization and Filter Optimization

```python
# Compare .loc[] vs .query() for filtering
start_time = time.time()
filter_loc = df.loc[df['age'] > 22]
end_time = time.time()
print(f"Time taken to filter using .loc[]: {end_time - start_time:.6f} seconds")

start_time = time.time()
filter_query = df.query('age > 22')
end_time = time.time()
print(f"Time taken to filter using .query(): {end_time - start_time:.6f} seconds")
```

```python
# Use .isin() for efficient filtering
multiple_majors = ['Computer Science', 'Arts']
start_time = time.time()
df_isin = df[df['major'].isin(multiple_majors)]
end_time = time.time()
print(f"\nTime taken to filter using .isin(): {end_time - start_time:.6f} seconds")
print(f"Rows filtered with .isin(): {len(df_isin)}")
```

```python
# Compare with multiple OR conditions 
start_time = time.time()
df_or = df[(df['major'] == 'Computer Science') | (df['major'] == 'Arts')]
end_time = time.time()
print(f"Time taken to filter using multiple OR conditions: {end_time - start_time:.6f} seconds")
print(f"Rows filtered with OR conditions: {len(df_or)}")
```

```python
# Chaining multiple Pandas operations (Filter, Groupby, Aggregation)
start_time = time.time()
summary_data = df.loc[df['enrollment_status'] == 'Active'].groupby('major').agg(
    average_gpa=('GPA', 'mean'),
    student_count=('GPA', 'count') 
).sort_values(by='average_gpa', ascending=False)
end_time = time.time()
print(f"\nTime taken for chained operations: {end_time - start_time:.6f} seconds")
print("\nChained Operations Output (Avg GPA and student count for Active students):")
print(summary_data)
```
