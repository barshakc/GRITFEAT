```python
import pandas as pd

# Load your CSV file
df = pd.read_csv("C:\Customer.csv")

# Convert 'dob' column from 'DD-MM-YYYY' to 'YYYY-MM-DD'
df['DOB'] = pd.to_datetime(df['DOB'], format='%d-%m-%Y').dt.strftime('%Y-%m-%d')

# Save to a new CSV file (or overwrite the old one if you prefer)
df.to_csv("customers_formatted.csv", index=False)

print("âœ… Date format conversion completed. File saved as 'customers_formatted.csv'")
```
